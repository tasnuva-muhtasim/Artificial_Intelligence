### Work with Train data

```python
train_data
```

Drop Unusual Value

```python
train_data = train_data.drop(['Cabin','PassengerId','Name','Ticket'],axis=1)
```

Handle Null Value

```python
train_data.isnull().sum()
```

```python
train_data['Age']=train_data['Age'].fillna(train_data['Age'].mean())
```

```python
train_data['Embarked']=train_data['Embarked'].fillna(train_data['Embarked'].mode()[0])
```

Encode Categorical value to numerical value

```python
train_data.info()
```

```python
train_data=pd.get_dummies(train_data,columns=['Sex','Embarked'])
```

```python
X = train_data.drop('Survived',axis = 1)
```

```python
Y = train_data['Survived']
```

```python
X.head()
```

```python
Y.head()
```

### Work with Test Data

```python
test_data
```

```python
test_data = test_data.drop(['Cabin','PassengerId','Name','Ticket'],axis=1)
```

```python
test_data.isnull().sum()
```

```python
test_data['Age']=test_data['Age'].fillna(test_data['Age'].mean())
```

```python
test_data['Fare']=test_data['Fare'].fillna(test_data['Fare'].mean())
```

```python
test_data.info()
```

```python
test_data=pd.get_dummies(test_data,columns=['Sex','Embarked'])
```

```python
test_data
```

### Data Splitting and Selecting the Best Algorithm

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
x_train,x_test,y_train,y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
accuracy_lr = accuracy_score(y_test, y_pred)
print(f"Accuracy : {accuracy_lr * 100:.2f}%")
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
accuracy_rf = accuracy_score(y_test, y_pred)
print(f"Accuracy : {accuracy_rf * 100:.2f}%")
```

### Applying the Best Algorithm and Creating a Submission FIle

```python
rf=RandomForestClassifier()
rf.fit(X,Y)
y_pred=rf.predict(test_data)
```
